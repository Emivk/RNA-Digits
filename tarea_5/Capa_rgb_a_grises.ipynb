```python
import tensorflow as tf
from tensorflow.keras.layers import Layer

class Capa_RGB_a_Grises(Layer):
    def __init__(self, ):
        super(Capa_RGB_a_Grises, self).__init__() #heredar propiedades de capa (supongo)

    def call(self, inputs):
        # Calcula la media ponderada de los canales de color para obtener la escala de grises
        imgris = tf.image.rgb_to_grayscale(inputs) #usamos el método de tf
        return imgris

```

```python
from PIL import Image
import tensorflow as tf
import numpy as np

# Ruta de la imagen
Rutaim = 'cat.10002.jpg'

# Cargar la imagen con PIL
imagen = Image.open(Rutaim) #leer la imagen y guardarla en la variable

imagen = imagen.resize((512, 512))  # Ajustar el tamaño de la imagen

# Convertir la imagen en un tensor necesitando numpy para darle forato a la func. de conversión
imagen = np.array(imagen) / 255.0  # Normalizar los valores de píxeles al rango [0, 1]
imagen = tf.convert_to_tensor(imagen, dtype=tf.float32)

# Añadir una dimensión para el batch
#imagen = imagen[tf.newaxis, ...]
```

```python
# creamos un objeto para la capa
img_a_gris=Capa_RGB_a_Grises()
# Aplicar la capa
imgris = img_a_gris(imagen) #ya esta la imagen a grises
```

```python
import matplotlib.pyplot as plt

# Suponiendo que `grayscale_image` contiene la imagen en escala de grises
# Asegúrate de que `grayscale_image` sea un tensor de TensorFlow con valores en el rango [0, 1]
# Puedes convertirlo a un array NumPy para la visualización

img_gris = imgris.numpy()  # Convierte el tensor a un array NumPy
plt.imshow(img_gris, cmap='gray')  # Muestra la imagen en escala de grises (para que no quede con colores azules y amarillos como lo hace matplot)
plt.axis('off')  # Oculta los ejes
plt.show()
```
